```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IMDB_Data.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.drop_duplicates(subset=['Movie_Name'],inplace=True)
```

```python
df.reset_index(inplace=True,drop = True)
df.to_csv('IMDB_Data_without_duplicates.csv',index = False)
```

```python
df.drop(['Movie_Metascore','Movie_Gross','Movie_Name','Movie_Date_Of_Release',
         'Movie_Rate','Movie_Length','Movie_Votes','Movie_Restriction'],axis=1,inplace=True)
```

```python
genres = ['Action', 'Sci-Fi', 'Comedy', 'Crime', 'Western','Family',
          'Drama', 'Horror', 'Romance', 'Thriller','Adventure']
```

```python
ser = df['Movie_Genres'].apply(lambda row : np.isin(genres, row.split(", ")).astype(int))
```

```python
res = pd.DataFrame(ser.to_list(), columns=genres)
```

```python
result = pd.concat([df.drop(['Movie_Genres'],axis = 1),res],axis =1)
result.to_csv('Movie_Description_and_Labels.csv', index = False)  
```
